```python
import pandas as pd

files = [
    "/content/drive/MyDrive/final exam 2024/Beans (dry).csv", "/content/drive/MyDrive/final exam 2024/Cassava.csv", "/content/drive/MyDrive/final exam 2024/Chili (red).csv",
    "/content/drive/MyDrive/final exam 2024/Maize.csv", "/content/drive/MyDrive/final exam 2024/Oranges (big size).csv",
    "/content/drive/MyDrive/final exam 2024/Peas (fresh).csv", "/content/drive/MyDrive/final exam 2024/Potatoes (Irish).csv",
    "/content/drive/MyDrive/final exam 2024/Sorghum.csv", "/content/drive/MyDrive/final exam 2024/Tomatoes.csv"
]

dataframes = {file.split('.')[0]: pd.read_csv(file) for file in files}

for key in dataframes:
    # Get the existing columns of the DataFrame
    existing_columns = dataframes[key].columns.tolist()

    # Replace the first column name with 'date'
    existing_columns[0] = 'date'

    # Replace the second column name with the key if it exists, otherwise keep the original name
    if len(existing_columns) > 1:  # Make sure there's a second column to rename
        existing_columns[1] = key

    # Assign the modified column names back to the DataFrame
    dataframes[key].columns = existing_columns
```

```python
# prompt: Convert Date Column to a Datetime Object

import pandas as pd

for key in dataframes:
    try:
        dataframes[key]['date'] = pd.to_datetime(dataframes[key]['date'])
    except ValueError:
        print(f"Error converting 'date' column in {key} DataFrame. Check the date format.")
        # Handle the error appropriately, e.g., skip the DataFrame or try a different date format
        continue #Skips to the next iteration if error occurs
    except KeyError:
      print(f"Error: 'date' column not found in {key} DataFrame.")
      continue
```

```python
# prompt: Combine All DataFrames

import pandas as pd

combined_df = pd.DataFrame()

for key in dataframes:
    try:
        # Concatenate dataframes vertically (row-wise)
        combined_df = pd.concat([combined_df, dataframes[key]], ignore_index=True)
    except KeyError as e:
        print(f"Error: Key '{e}' not found in dataframes dictionary.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Display information about the combined dataframe
print(combined_df.info())

# Now combined_df holds all the data from individual dataframes.
# You can perform operations on this combined dataframe.
```

```python
# prompt: Handle Missing Dates

# Assuming you have already executed the code you provided to create the 'dataframes' dictionary

# Function to handle missing dates in the combined dataframe
def handle_missing_dates(df):
    # 1. Identify the date range
    min_date = df['date'].min()
    max_date = df['date'].max()
    date_range = pd.date_range(start=min_date, end=max_date, freq='D')

    # 2. Create a complete date index
    all_dates = pd.DataFrame({'date': date_range})

    # 3. Merge with the original DataFrame on 'date' column
    merged_df = pd.merge(all_dates, df, on='date', how='left')

    # 4. Fill NaN values for relevant columns
    # Replace 'column1', 'column2', etc., with the actual names of the columns with data
    # Choose an appropriate method for filling NaN values. ffill (forward fill) is used here
    for col in merged_df.columns[1:]:  # Skip the date column
        if merged_df[col].dtype != 'object' and merged_df[col].dtype != 'datetime64[ns]': # Do not fill object or datetime types
           merged_df[col] = merged_df[col].fillna(method='ffill')

    return merged_df


# Handle missing dates in the combined DataFrame
combined_df = handle_missing_dates(combined_df)

# Display information about the combined dataframe
print(combined_df.info())
```

# Forecasting for the Next 6 **Months**

```python
# Forecasting for the Next 6 Months

from statsmodels.tsa.arima.model import ARIMA
from datetime import timedelta
import pandas as pd

# Assuming 'combined_df' is your original DataFrame with the 'date' column
# and 'numeric_df' contains only numeric columns

forecast_dfs = {}  # Store forecasts for each product

for product in numeric_df.columns:
    # Skip the adm1_id and mkt_id columns as well
    if product not in ['adm1_id', 'mkt_id', 'cm_id', 'cur_id', 'pt_id', 'um_id']:

        # Prepare the time series data using the original 'combined_df'
        product_ts = combined_df[['date', product]].set_index('date')
        product_ts = product_ts.fillna(method='ffill')

        # Fit the ARIMA model
        try:
            model = ARIMA(product_ts, order=(5, 1, 0)) # You can adjust the order
            model_fit = model.fit()

            # Forecast for the next 6 months (180 days)
            forecast_steps = 180  # 6 months * 30 days/month
            forecast = model_fit.forecast(steps=forecast_steps)

            # Create a date range for the forecast
            last_date = product_ts.index.max()
            future_dates = pd.date_range(start=last_date + timedelta(days=1), periods=forecast_steps)

            # Create a DataFrame for the forecast
            forecast_df = pd.DataFrame({'date': future_dates, product: forecast})
            forecast_dfs[product] = forecast_df
            print(f'Forecast for {product} generated successfully.')

        except Exception as e:
            print(f"Error forecasting for {product}: {e}")


# Combine all forecasts into a single DataFrame
combined_forecasts = pd.DataFrame()

for product, df in forecast_dfs.items():
    if combined_forecasts.empty:
        combined_forecasts = df
    else:
        combined_forecasts = pd.merge(combined_forecasts, df, on='date', how='outer')

# Display or save the combined forecasts
print(combined_forecasts.head())


#Plotting the forecasts
for product in numeric_df.columns:
    # Skip the adm1_id and mkt_id columns as well
    if product not in ['adm1_id', 'mkt_id', 'cm_id', 'cur_id', 'pt_id', 'um_id']:
        plt.figure(figsize=(10, 6))
        plt.plot(combined_df['date'], combined_df[product], label='Actual')  # Use combined_df for actual data
        plt.plot(combined_forecasts['date'], combined_forecasts[product], label='Forecast')
        plt.xlabel('Date')
        plt.ylabel(f'{product}')
        plt.title(f'Forecast of {product}')
        plt.legend()
        plt.show()
```
